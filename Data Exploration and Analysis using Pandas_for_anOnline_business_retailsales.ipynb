**Data Dictionary to Understand what each column represents**:


`Total Orders`: ***The total number of orders placed.***

`Type`: ***Integer***

`Gross Sales`:***The total sales revenue generated before any deductions (discounts, returns, etc.).***

`Type`:***Numerical***


`Discounts`:***The total amount deducted from the gross sales due to discounts applied***.

`Type`: ***Numerical***


`Returns`: ***The total amount of sales returned by customers.***

`Type`: ***Numerical***


`Net Sales`: ***The total sales revenue after deducting discounts and returns from gross sales.
Calculation: Gross Sales - Discounts - Returns***

`Type`: ***Numerical***


`Shipping`: ***The total cost associated with shipping orders.***

***Type: Numerical***


`Total Sales`: ***The total revenue generated, including net sales and shipping charges.***

`Calculation`: ***Net Sales + Shipping***

# **`DENSITY PLOT FOR NUMERICAL COLUMNS`**

```python
for column in num_columns:
    plt.figure(figsize=(20,8))
    sb.kdeplot(business_retailsales_df[column],color='blue')
    plt.title(f'Density plot of{column}')
    plt.ylabel(column)
    plt.show()
```

**`Box plot of Numerical columns`**

```python
for column in num_columns:
    plt.figure(figsize=(20,8))
    plt.boxplot(business_retailsales_df[column],patch_artist=True,boxprops=dict(facecolor='lightblue',color='blue'), medianprops=dict(color='red',linewidth=5),whiskerprops=dict(color='green',linewidth=1.6,linestyle='--'),capprops=dict(color='purple',linewidth =1.5),flierprops=dict(marker='o',markersize=8,markerfacecolor='yellow',markeredgecolor='black'),vert=False)
    plt.title(f'Box plot of{column}')
    plt.ylabel(column)
    plt.grid(True, linestyle= '-', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```
